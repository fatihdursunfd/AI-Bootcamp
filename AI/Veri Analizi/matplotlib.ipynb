```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
plt.style.use('seaborn-white')
plt.figure(figsize=(12,6))
```

```python
# line graph

x = np.array([i for i in range(-100,100)])
y = np.power(x,3)

plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(0,10,1000)
y = np.sin(x)

plt.plot(x, y, c="#C65D7B",marker=".")
plt.ylabel("y")
plt.xlabel("x")
plt.show()
```

```python
x = np.array([i for i in range(-9,10)])
y1 = np.power(x,2)
y2 = np.power(x,2) + 3*x + 15 

plt.plot(x, y1, "b-", label='$y = x^2 $')
plt.plot(x, y2, "r-", label='$y = x^2 + 3x + 15$')

plt.legend()
plt.tight_layout()

plt.show()
```

```python
x = np.linspace(0,10,2000)

plt.plot(x , np.sin(x) , label='y = sin(x)' )
plt.plot(x , np.cos(x) , label='y = cos(x)' )

plt.xlabel(r'$X$' , fontsize = 18)
plt.ylabel(r'$Y$' , fontsize = 18)
plt.title("$sin(x) $ $ & $ $ cos(x)$" , fontsize=18)

plt.legend(loc = "lower left")
plt.show()
```

```python
x = np.linspace(0,10,2000)

plt.plot(x , np.sin(x) , label='y = sin(x)' )
plt.plot(x , np.cos(x) , label='y = cos(x)' )

plt.fill_between(x , 0 , np.sin(x))
plt.fill_between(x , 0 , np.cos(x))

plt.xlabel(r'$X$' , fontsize = 18)
plt.ylabel(r'$Y$' , fontsize = 18)
plt.title("$sin(x) $ $ & $ $ cos(x)$" , fontsize=18)

plt.legend(loc = "lower left")
plt.show()
```

```python
# stacked bar

x1 = ["A1" , "A2" , "A3", "A4"]

y1 = [11,21,15,26]
y2 = [2,7,14,5]

plt.figure(figsize=(5,7))

plt.bar(x1, y1, label="sold tickets" , width=0.5, color="#E9896A")
plt.bar(x1, y2, label="ticket in sale" , width=0.5, bottom=y1, color="#387C6D")

plt.legend()

plt.show()
```

```python
# horizontal stack bar

x1 = ["A1" , "A2" , "A3", "A4"]

y1 = [11,21,15,26]
y2 = [5,7,14,5]

plt.figure(figsize=(5,7))

plt.barh(x1, y1, label="sold tickets" ,height=0.3, color="#6B011F")
plt.barh(x1, y2, label="ticket in sale" , height=0.3, left=y1 , color="#290149")

plt.legend()

plt.show()
```

```python
# pie chart

plt.figure(figsize=(9,9))

area = [48, 30, 20, 15]
labels = ["a", "b", "c", "d"]
colors = ["#BB6464","#CDB699", "#C3DBD9", "#C8F2EF"]
plt.pie(area, labels=labels, colors=colors, startangle=61 , textprops={'fontsize': 25})
plt.show()
```

```python
plt.figure(figsize=(9,9))

area = [48, 30, 20, 15]
labels = ["low", "medium", "high", "critical"]

total = np.sum(area)

def val_per(x):
    return '{:.2f}%\n({:.0f})'.format(x, x * total / 100)


colors = ["#D3ECA7","#A1B57D", "#B33030", "#19282F"]
plt.pie(area, labels=labels, colors=colors, startangle=0, shadow='true', autopct=val_per , textprops={'fontsize': 16})
plt.show()
```

```python
# donut plot

fig = plt.figure(figsize=(9,9))

# set background color
fig.patch.set_facecolor("#B4B897")

area = [48, 30, 20, 15]
labels = ["low", "medium", "high", "critical"]
colors = ["#D3ECA7","#A1B57D", "#B33030", "#19282F"]

plt.pie(area , labels=labels , colors=colors , startangle=0 , shadow='true', autopct='%1.1f%%'  ,textprops={'fontsize': 16})

my_circle = plt.Circle((0,0), 0.45, color="#B4B897")
p = plt.gcf()
p.gca().add_artist(my_circle)

plt.show()
```

```python
fig = plt.figure(figsize=(9,9))

# set background color
fig.patch.set_facecolor("#B4B897")

area = [48, 30, 20, 15]
labels = ["low", "medium", "high", "critical"]
colors = ["#D3ECA7","#A1B57D", "#B33030", "#19282F"]

total = np.sum(area)
def val_per(x):
    return '{:.2f}%\n({:.0f})'.format(x, x * total / 100)

plt.pie(area , labels=labels , colors=colors , startangle=0 , shadow='true', autopct=val_per , pctdistance=0.71, explode=[0,0.2,0,0.2] , textprops={'fontsize': 16})

my_circle = plt.Circle((0,0), 0.45, color="#B4B897")
p = plt.gcf()
p.gca().add_artist(my_circle)

plt.show()
```

```python
fig = plt.figure(figsize=(20,6))
fig.patch.set_facecolor("#B4B897")

area = [61, 16, 34, 81]
priority = ["Low", "Medium", "High", "Critical"]
status = ["In progress", "Next Up", "Completed", "Cancelled"]
company = ["IBM", "ASF Studio", "Microsoft", "Samsung"]

colors = ["#5F939A","#5F939A","#A0937D","#F2EDD7",]

plt.subplot(1,3,1)
plt.pie(area, labels=priority, colors=colors, startangle=0)

plt.subplot(1,3,2)
plt.pie(area, labels=status, colors=colors, startangle=0)

plt.subplot(1,3,3)
plt.pie(area, labels=company, colors=colors, startangle=0)

plt.show()
```

```python
fig , axs = plt.subplots(2,3, figsize=(20,12))

fig.patch.set_facecolor("#B4B897")

area = [61, 16, 34, 81]
priority = ["Low", "Medium", "High", "Critical"]
status = ["In progress", "Next Up", "Completed", "Cancelled"]
company = ["IBM", "ASF Studio", "Microsoft", "Samsung"]
employee = ["Fatih", "Ali", "Kadir", "Jack"]
deadline = ["1 Month", "1 Year", "10 Day", "10 Week"]
price = ["1.000$", "1Million$", "10.000$", "10Million$"]

colors = ["#5F939A","#5F939A","#A0937D","#F2EDD7",]

axs[0, 0].pie(area, labels=priority, colors=colors, startangle=0)
axs[0, 1].pie(area, labels=status, colors=colors, startangle=0)
axs[0, 2].pie(area, labels=company, colors=colors, startangle=0)
axs[1, 0].pie(area, labels=employee, colors=colors, startangle=0)
axs[1, 1].pie(area, labels=deadline, colors=colors, startangle=0)
axs[1, 2].pie(area, labels=price, colors=colors, startangle=0)

plt.show()
```
