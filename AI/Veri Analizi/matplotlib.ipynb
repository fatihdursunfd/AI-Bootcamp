```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
plt.style.use('seaborn-white')
plt.figure(figsize=(9,5))
```

```python
# line graph

x = np.array([i for i in range(-100,100)])
y = np.power(x,3)

plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(0,10,1000)
y = np.sin(x)

plt.plot(x, y, c="#C65D7B",marker=".")
plt.ylabel("y")
plt.xlabel("x")
plt.show()
```

```python
x = np.array([i for i in range(-9,10)])
y1 = np.power(x,2)
y2 = np.power(x,2) + 3*x + 15 

plt.plot(x, y1, "b-", label='$y = x^2 $')
plt.plot(x, y2, "r-", label='$y = x^2 + 3x + 15$')

plt.legend()
plt.tight_layout()

plt.show()
```

```python
x = np.linspace(0,10,2000)

plt.plot(x , np.sin(x) , label='y = sin(x)' )
plt.plot(x , np.cos(x) , label='y = cos(x)' )

plt.xlabel(r'$X$' , fontsize = 18)
plt.ylabel(r'$Y$' , fontsize = 18)
plt.title("$sin(x) $ $ & $ $ cos(x)$" , fontsize=18)

plt.legend(loc = "lower left")
plt.show()
```

```python
x = np.linspace(0,10,2000)

plt.plot(x , np.sin(x) , label='y = sin(x)' )
plt.plot(x , np.cos(x) , label='y = cos(x)' )

plt.fill_between(x , 0 , np.sin(x))
plt.fill_between(x , 0 , np.cos(x))

plt.xlabel(r'$X$' , fontsize = 18)
plt.ylabel(r'$Y$' , fontsize = 18)
plt.title("$sin(x) $ $ & $ $ cos(x)$" , fontsize=18)

plt.legend(loc = "lower left")
plt.show()
```

```python
plt.figure(figsize=(14,3))

x = np.linspace(0,10,100)

y1 = np.sin(x)
y2 = np.cos(x)

plt.subplot(1,2,1)
plt.plot(x,y1)

plt.subplot(1,2,2)
plt.plot(x,y2)

plt.show()
```

```python
plt.figure(figsize=(12,6))

x = np.linspace(0,10,100)

y1 = np.sin(x)
y2 = np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y1)

plt.subplot(2,1,2)
plt.plot(x,y2)

plt.show()
```

```python
# bar graphs

id1 =np.arange(0,10)
score = np.arange(0,100,10)

ax = plt.axes()
ax.set_facecolor("#8D8DAA")

plt.xlabel(r'$Student ID$')
plt.ylabel(r'$Score$')
plt.bar(id1, score , color="#DFDFDE")
plt.show()
```

```python
x1 = [1,3,5,7]
x2 = [2,4,6,8]

y1 = [7,17,1,11]
y2 = [16,8,8,4]

plt.figure(figsize=(8,6))

ax = plt.axes()
ax.set_facecolor("#F1E9E5")

plt.xlabel(r'$X$' , fontsize = 14)
plt.ylabel(r'$Y$' , fontsize = 14)

plt.bar(x1, y1 , label='First', color="#368B85")
plt.bar(x2, y2 , label='Second', color="#464660")

plt.show()
```

```python
Age = [23, 25, 4, 45, 16]
Name = ["Fatih" , "Kadir" , "Taha", "Ali", "Veli"]

plt.barh(Name, Age, color="#5F939A")
plt.show()
```

```python
num1 = np.array([i for i in range(1,10,2)])
num2 = np.array([i for i in range(0,9,2)])

plt.figure(figsize=(8,4))

plt.bar(num1 , num1**2 , width=0.3 , color="#8E9775")
plt.bar(num2 , num2**2 , width=0.3 , color="#E28F83")

plt.show()
```

```python
# stacked bar

x1 = ["A1" , "A2" , "A3", "A4"]

y1 = [11,21,15,26]
y2 = [2,7,14,5]

plt.figure(figsize=(5,7))

plt.bar(x1, y1, label="sold tickets" , width=0.5, color="#E9896A")
plt.bar(x1, y2, label="ticket in sale" , width=0.5, bottom=y1, color="#387C6D")

plt.legend()

plt.show()
```

```python
# horizontal stack bar

x1 = ["A1" , "A2" , "A3", "A4"]

y1 = [11,21,15,26]
y2 = [5,7,14,5]

plt.figure(figsize=(5,7))

plt.barh(x1, y1, label="sold tickets" ,height=0.3, color="#6B011F")
plt.barh(x1, y2, label="ticket in sale" , height=0.3, left=y1 , color="#290149")

plt.legend()

plt.show()
```
